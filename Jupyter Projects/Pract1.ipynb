Label Encoding-                         
 Import pandas and sklearn library for preprocessing

```python
from sklearn import preprocessing
```

Observe the unique values for the Position column.

```python
df['Position'].unique()
```

define label_encoder object knows how to understand word labels

```python
label_encoder=preprocessing.LabelEncoder()
```

Encode labels in column 'Position'

```python
df['Position']=label_encoder.fit_transform(df['Position'])
```
