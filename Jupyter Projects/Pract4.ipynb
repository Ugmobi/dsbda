Create a Linear Regression model using python to predict home prise using
Boston Housing dataset.

Import Required Libraries

```python
import pandas as pd
```

Load the Dataset

```python
df=pd.read_csv("BostonHousing.csv")
```

```python
df
```

head() method returns a specified number of rows, string from the top.

```python
df.head()
```

Import data  using drop


```python
x=df.drop('medv',axis=1)
y=df['medv']
```

used to fetch the dimensions of Pandas

```python
x.shape
```

Split arrays or matrices into random train and test subsets.

```python
from sklearn.model_selection import train_test_split
```

X_train and y_train sets are used for training and fitting the model.

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,random_state=0,test_size=0.20)
```

```python
x_train
```

head() method returns a specified number of rows, string from the top.

```python
x_train.head()
```

shape enables us to obtain the shape of a DataFrame

```python
x_train.shape
```

minimize the residual sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation

```python
from sklearn.linear_model import LinearRegression
```

used to predict the value of a variable based on the value of another variable

```python
regressor=LinearRegression()
```

Learn and estimate the parameters of the transformation

```python
regressor.fit(x_train,y_train)
```

```python
LinearRegression()
```

used to estimate the coefficients for the linear regression problem

```python
regressor.coef_
```

shows the point where the estimated regression line crosses the 𝑦 axis

```python
regressor.intercept_
```

given a trained model, predict the label of a new set of data.

```python
y_pred=regressor.predict(x_test)
```

the data predicted (calculated, output) by your model

```python
y_pred.shape
```

contains two-dimensional data and its corresponding labels

```python
result=pd.DataFrame({'Actual':y_test,'Producted':y_pred})
```

```python
result
```

Return a Series/DataFrame with absolute numeric value of each element

```python
residual_error=abs(y_test-y_pred)
```

```python
residual_error
```

```python
sum(residual_error)/len(residual_error)
```

Creates two local variables, total and count that are used to compute the mean absolute error

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_pred,y_test)
```

metrics let you assess the quality of your predictions

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

score as a shorthand to calculate accuracy since it is such a common metric.

```python
regressor.score(x_test,y_test)
```

r2_score regression score function

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
